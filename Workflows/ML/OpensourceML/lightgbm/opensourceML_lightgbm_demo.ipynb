<b style = 'font-size:28px;font-family:Arial;color:#E37C4D'>Introduction</b><br>
OpensourceML enables users to run opensource libraries in Teradata Vantage, without pulling the data to client.</br>
We already exposed an interface object `td_sklearn` on which users can run scikit-learn classes and functions.</br>
Similarly, we are exposing `td_lightgbm` an interface object for lightgbm package.

<b style = 'font-size:28px;font-family:Arial;color:#E37C4D'>Import the required libraries</b>

```python
# Importing required libraries
import getpass
import random
from teradataml import *
from teradataml.opensource import td_lightgbm
```

<b style = 'font-size:28px;font-family:Arial;color:#E37C4D'>Connect to Vantage</b>

```python
# Create the connection.
td_context = create_context(host=getpass.getpass("Hostname: "), username=getpass.getpass("Username: "), password=getpass.getpass("Password: "))
```

<b style = 'font-size:28px;font-family:Arial;color:#E37C4D'>Loading Dataset</b>

```python
# Load the example dataset.
load_example_data("openml", ["multi_model_classification", "multi_model_regression"])
```

<b style = 'font-size:20px;font-family:Arial;color:#E37C4D'>Classification Dataset</b>

```python
df_train_classif = DataFrame('multi_model_classification')
df_train_classif
```

```python
# Create required classification DataFrames.
df_x_classif = df_train_classif.select(["col1", "col2", "col3", "col4"])
df_y_classif = df_train_classif.select("label")
```

<b style = 'font-size:20px;font-family:Arial;color:#E37C4D'>Regression Dataset</b>

```python
df_train_reg = DataFrame('multi_model_regression')
df_train_reg
```

```python
# Create required regression DataFrames.
df_x_reg = df_train_reg.select(["col1", "col2", "col3", "col4"])
df_y_reg = df_train_reg.select("label")
```

<b style = 'font-size:28px;font-family:Arial;color:#E37C4D'>Single Model case</b>

<b style = 'font-size:20px;font-family:Arial;color:#E37C4D'>Creating Dataset Object</b>

```python
obj_s = td_lightgbm.Dataset(df_x_classif, df_y_classif, silent=True, free_raw_data=False)
obj_s
```

```python
obj_s1 = td_lightgbm.Dataset(df_x_classif, df_y_classif, silent=True, free_raw_data=False)
obj_s1
```

```python
# construct() and run get_data()
obj_s.construct()
```

```python
obj_s.get_data()
```

```python
# subset
obj_s_4 = obj_s.subset([1,2,5])
obj_s_4.construct()
obj_s_4.get_data()
```

```python
# save_binary(): Save to both client and Vantage. Added another arguement "save_in_vantage".
obj_s.save_binary("save_in_vantage_also_single.model", save_in_vantage=True)
```

<b style = 'font-size:20px;font-family:Arial;color:#E37C4D'>Run Training APIs</b>

<b style = 'font-size:16px;font-family:Arial;color:#E37C4D'>train()</b>

```python
rec = {} 
```

```python
# Training With valid_sets and callbacks argument.
opt_tr = td_lightgbm.train(params={}, train_set=obj_s1, num_boost_round=30,
                           callbacks=[td_lightgbm.record_evaluation(rec), td_lightgbm.early_stopping(3)],
                           valid_sets=[obj_s1])
opt_tr
```

```python
opt_tr.record_evaluation_result
```

<b style = 'font-size:16px;font-family:Arial;color:#E37C4D'>cv()</b>

```python
# Training without return_cvbooster argument.
opt_cv = td_lightgbm.cv(params={}, train_set=obj_s1, callbacks=[td_lightgbm.early_stopping(5)])
opt_cv
```

`label` argument is added to `predict()` by teradataml for the function to 
return label column in output DataFrame.

```python
# With pred_contrib=True (returns more columns) and without label argument in
# output also.
opt_pred = opt_tr.predict(data=df_x_classif, num_iteration=20, pred_contrib=True)
opt_pred
```

```python
# With pred_contrib=False (returns only one column - predicted value) and with label argument.
opt_pred = opt_tr.predict(data=df_x_classif, label=df_y_classif, num_iteration=20)
opt_pred
```

<b style = 'font-size:28px;font-family:Arial;color:#E37C4D'>Supportability</b>

<b>Note:</b></br>
- Some functions in Dataset class like `set_weight()`, `set_label()` etc are not supported.
- `cv()` with `return_cvbooster` argument is not yet supported.
- Some functions of Booster class are not supported.
- All Booster functions that are supported with single model execution are supported on multi-model except `model_from_string()`.
- To use `LGBMRanker` in multi model case, all partitions should have same number of samples.
- Most of the limitations with OpensourceML like callable object exists even for LightGBM.

<b>Please refer User guide and Supportability Matrix that is shared with teradataml 20.00.00.03 to understand which functions are supported and which are not.</b>
