```python
# Create an Array object with integer values and assign it to a new column.
from teradataml.dataframe.array import Array
from teradatasqlalchemy.types import ARRAY_INTEGER

arr_obj = Array((1, 2, 3, 4, 5), ARRAY_INTEGER('[5]'))
df = df_arrays.assign(new_array = arr_obj)
print_variables(df)
```

```python
# Create an Array object with VARCHAR values.
from teradatasqlalchemy.types import ARRAY_VARCHAR

arr_varchar = Array(('hello', 'world', 'test'), ARRAY_VARCHAR('[3]'))
df = df_arrays.assign(string_array = arr_varchar)
print_variables(df)
```

### DataFrame.unnest() Method

```python
# Explode the array column 'arr1' without key column and ordinality.
res = df_arrays.unnest(array_col="arr1")
print_variables(res)
```

```python
# Explode the array column 'arr3' with key column 'id' and ordinality.
df = df_arrays.unnest(array_col=df_arrays.arr3, key_col="id", ordinality=True)
print_variables(df)
```
